```python
import pandas as pd

# Đọc dữ liệu từ file CSV
df = pd.read_csv('cleanconditionady.csv')  # Thay 'your_file.csv' bằng tên file của bạn

# Tính số phần mà bạn muốn chia
num_splits = 5
rows_per_split = len(df) // num_splits  # Số dòng cho mỗi bảng nhỏ

# Chia bảng thành các phần
dfs = [df.iloc[i*rows_per_split:(i+1)*rows_per_split] for i in range(num_splits)]

# Nếu số dòng của DataFrame không chia hết cho 5, bạn có thể phân bổ phần còn lại vào các bảng cuối cùng
remaining_rows = len(df) % num_splits
if remaining_rows > 0:
    for i in range(remaining_rows):
        # Thêm các dòng còn lại vào bảng nhỏ từ đầu
        dfs[i] = pd.concat([dfs[i], df.iloc[num_splits*rows_per_split + i:num_splits*rows_per_split + i+1]])

# Kiểm tra kết quả, in ra 5 dòng đầu tiên của mỗi bảng để xem dữ liệu
for i, small_df in enumerate(dfs):
    print(f"DataFrame {i+1}:")
    print(small_df.head())  # In ra 5 dòng đầu tiên của mỗi bảng
    print()

# Lưu các bảng nhỏ vào các file CSV riêng biệt
for i, small_df in enumerate(dfs):
    small_df.to_csv(f"split_{i+1}.csv", index=False, encoding='utf-8')  # Lưu các bảng vào các file CSV

print("Chia bảng thành công và lưu vào các file CSV.")
```
